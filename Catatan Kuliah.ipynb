```python
#Quick Sort Pivot di akhir
def quicksortawal(a):
    quicksort(a,0,len(a)-1)
    
    
def quicksort(a,start,end):
    if start<end:
        pindex = partition(a,start,end)
        quicksort(a,start,pindex-1)
        print(a)
        quicksort(a,pindex+1,end)
        print(a)

        
def partition(a,start,end):
    pivot = a[end]
    pindex = start
    for i in range(start,end):
        if (a[i]<=pivot):
            a[i],a[pindex] = a[pindex],a[i]
            pindex=pindex+1
    a[pindex],a[end]=a[end],a[pindex]
    return pindex

a=[100,6,500,4,56,2,1,0]
print(a)
quicksortawal(a)
print(a)
```

```python
#Quick Sort Pivot di awal
def quicksortawal(a):
    quicksort(a,0,len(a)-1)
    
    
def quicksort(a,start,end):
    if start<end:
        pindex = partition(a,start,end)
        quicksort(a,start,pindex-1)
        print(a)
        quicksort(a,pindex+1,end)
        print(a)

        
def partition(a,start,end):
    pivot = a[start]
    pindex = end
    for i in range(end,start,-1):
        if (a[i]>=pivot):
            a[i],a[pindex] = a[pindex],a[i]
            pindex=pindex-1
    a[pindex],a[start]=a[start],a[pindex]
    return pindex

a=[100,6,500,4,56,2,1,0]
print(a)
quicksortawal(a)
print(a)
```

```python
#MERGE SORT
def mergesort(a):
    n = len(a)
    if (n<2):
        return
    mid = len(a)//2
    left = a[:mid]
    right = a[mid:]
    mergesort(left)
    mergesort(right)
    i=0
    j=0
    k=0
    
    while i<len(left) and j <len(right):
        if left[i] < right[j]:
            a[k]=left[i]
            i=i+1
            
        else:
            a[k]=right[j]
            j=j+1
        k=k+1
        
    while j <len(right):
        a[k]=right[j]
        j=j+1
        k=k+1
    while i <len(left):
        a[k]=left[i]
        i=i+1
        k=k+1
        
a=[52,26,93,17,99,77,31,44,55,20,100]
mergesort(a)
print(a)
```

```python
#SHELL SORT
def shellsort(a):
    sub=len(a)//2
    while sub > 0:
        
        for awal in range(sub):
            print(sub)
            gapsort(a,awal,sub)
        sub=sub//2
        
def gapsort(a,start,gap):
    for i in range(start+gap,len(a),gap):
        
        value = a[i]
        position = i
        
        while position>=gap and a[position-gap]>value:
            a[position]=a[position-gap]
            position=position-gap
            
        a[position]=value
        print(a)
        
a=[35,33,42,10,14,19,27,44]
print(a)
shellsort(a)
print(a)
```

```python
#STACK 1
class stack:
    def __init__(self):
        self.items = []
        
    def isempty(self):
        return self.items==[]
        
    def push(self,item):
        self.items.append(item)
            
    def pop(self):
        return self.items.pop()
        
    def peek(self):
        return self.items[len(self.items)-1]
        
    def size(self):
        return len(self.items)
        
s=stack()
s.push ("hello")
s.push ("true")
print(s.isempty)
print(s.peek())
while not s.isempty():
    print(s.pop())
```

```python
#QUEUE
class qeue:
    def __init__(self):
        self.items=[]
        
    def isempty(self):
        return self.items==[]
    
    def en(self,item):
        self.items.insert(0,item)
        
    def de(self):
        return self.items.pop()
    
    def size(self):
        return len(self.items)
    
q=qeue()
q.en(4)
q.en('dog')
q.en(True)
print(q.size())
print(q.de())
```

```python
from pythonds.basic.stack import Stack

def fix(post):
    operandStack = Stack()
    tokenList = post()
    
    for token in tokenList:
        if token in "0123456789":
            operandStack.push(int(token))
        else:
            operand2 = operandStack.pop()
            operand1 = operandStack.pop()
            result = doMath(token,operand1,operand2)
            operandStack.push(result)
    return operandStack.pop()

def doMath(op,op1,op2):
    if op =="*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    else:
        return op1 - op2
print(fix("7 8 + 3 2 + /"))
```
